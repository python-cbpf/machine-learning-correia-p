```python
import numpy as np
import csv
import math
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop
```

```python
##Importing the data

x = []
y = []

with open('pseudo_data_square.csv', newline = '') as f:
    reader = csv.reader(f, delimiter = '\t')
    for line in reader:
        x.append((np.array(line).astype(np.float)[:2]).tolist())
        y.append((np.array(line).astype(np.float)[2]).tolist())
```

```python
model = Sequential()
```

```python
model.add(Dense(1, activation='sigmoid', use_bias=True, input_dim=2))
model.add(Dense(1, activation='sigmoid', use_bias=True, input_dim=2))
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x[:250], y[:250], epochs=1000, batch_size=32)
```

```python
model.evaluate(x[250:], y[250:], batch_size=32)
```

```python
plt.hist(model.predict(x[250:]), bins=100)
plt.show()
```

```python
y_pred = model.predict(x[250:])
y_true = y[250:]

loss = np.zeros(len(y_pred))
for i in range(len(y_pred)):
    loss[i] = (1/2)*(y_pred[i]-y_true[i])**2

plt.hist(loss, bins = 100)
plt.show()
```
